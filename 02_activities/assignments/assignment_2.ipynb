#### Starter Code for Question 1

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val = 0, left = None, right = None):
#         self.val = val
#         self.left = left
#         self.right = right
from collections import deque

class TreeNode(object):
     def __init__(self, val = 0, left = None, right = None):
         self.val = val
         self.left = left
         self.right = right

def is_duplicate(root: TreeNode) -> int:
    if not root:
        return -1
    
    seen = set()
    queue = deque([root])

    while queue:
        node = queue.popleft()

        if node.val in seen:
            return node.val
        seen.add(node.val)
        
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    return -1
```

>It is Level Order Traversal (Breadth-First-Search, BFS).

```python
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(2)
root.left.left = TreeNode(3)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

is_duplicate(root)
```

>The problem is to check whether the brackets in the string are properly matched and correctly ordered. A string is considered valid if every openning bracket has a matching closing bracket of the same type, and the brackets are closed in the correct sequence. 


-   Create 1 new example that demonstrates you understand the problem. Trace/walkthrough 1 example that your partner made and explain it.


>Ex1:
>Input: s = "(([]))"
>Output: True

>Ex2:
>Input: s = "{[(])}"
>Output: False


-   Copy the solution your partner wrote. 


```python
# Dictionary to define the accepted bracket pairs
matches = {")": "(", "]": "[", "}": "{"}

def is_valid_brackets(s): 
    opening_brackets = []  # empty stack to store unmatched opening brackets
    for i in s:            # loop for each character in the string                          
        if i in "([{": 
            opening_brackets.append(i)  # open bracket characters are pushed onto the stack
        elif opening_brackets == []:    # if a close bracket character is encountered with nothing in the stack, False
            return False
        elif matches[i] == opening_brackets[-1]: # if a close bracket characters matches the last stacked item, True, and the last stacked item is popped
            opening_brackets.pop()
        else: 
            return False
    return opening_brackets == [] # if the stack is empty at the end, True, False otherwise
    
```


-   Explain why their solution works in your own words.


>This solution used a stack to track openning brackets. It pushes openning brackets onto the stack, and for each closing bracket it checks whether it matches the most recent openning bracket. If doesn't match or the stack is empty, it returns false. In the end, the string is valid only if the stack is empty, meaning all brackets were correctly matched and closed in order.


-   Explain the problem’s time and space complexity in your own words.


>The algorithm runs O(n) time complexity, where n is the length of the input of string. Because it has n iteraitions for each element in the string.
>For the space completity, it is still O(n) because the worst case os tje stacl grows the same size of the input size.
